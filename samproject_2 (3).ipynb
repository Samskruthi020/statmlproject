```python
#linear bootstrap values graph
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate synthetic data (replace with your own dataset)
np.random.seed(0)
X = np.random.rand(100, 1)
y = 2 * X + 1 + 0.1 * np.random.randn(100, 1)

# Number of bootstrap samples
num_samples = 1000

# Initialize arrays to store coefficients
intercepts = np.zeros(num_samples)
slopes = np.zeros(num_samples)

# Perform bootstrapping
for i in range(num_samples):
    indices = np.random.choice(len(X), len(X), replace=True)
    X_bootstrap = X[indices]
    y_bootstrap = y[indices]

    # Fit a linear regression model to the bootstrap sample
    model = LinearRegression()
    model.fit(X_bootstrap, y_bootstrap)

    # Store the intercept and slope coefficients
    intercepts[i] = model.intercept_[0]
    slopes[i] = model.coef_[0][0]

# Visualize the distribution of coefficients
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(intercepts, bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Intercepts')
plt.xlabel('Intercept')
plt.ylabel('Frequency')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.hist(slopes, bins=30, color='lightcoral', edgecolor='black')
plt.title('Distribution of Slopes')
plt.xlabel('Slope')
plt.ylabel('Frequency')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
#linear bootstrap
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Independent variable
y = np.array([2, 4, 5, 4, 5])  # Dependent variable

# Fit a linear regression model to the original data
original_model = LinearRegression()
original_model.fit(X, y)

# Get the intercept and slope from the original model
original_intercept = original_model.intercept_
original_slope = original_model.coef_[0]

# Print the original coefficients
print("Original Intercept:", original_intercept)
print("Original Slope:", original_slope)

# Perform bootstrapping
n_bootstraps = 1000  # Number of bootstrap samples
intercepts = []
slopes = []

for _ in range(n_bootstraps):
    # Resample with replacement
    indices = np.random.choice(len(X), len(X), replace=True)
    X_bootstrap = X[indices]
    y_bootstrap = y[indices]

    # Fit a linear regression model to the bootstrap sample
    model = LinearRegression()
    model.fit(X_bootstrap, y_bootstrap)

    # Get the intercept and slope from the bootstrap model
    intercepts.append(model.intercept_)
    slopes.append(model.coef_[0])

# Calculate the means and standard deviations of the bootstrapped coefficients
mean_intercept = np.mean(intercepts)
mean_slope = np.mean(slopes)

std_intercept = np.std(intercepts)
std_slope = np.std(slopes)

# Print the means and standard deviations of the bootstrapped coefficients
print("Mean of Bootstrapped Intercepts:", mean_intercept)
print("Mean of Bootstrapped Slopes:", mean_slope)
print("Standard Deviation of Bootstrapped Intercepts:", std_intercept)
print("Standard Deviation of Bootstrapped Slopes:", std_slope)
```

```python
#svm bootstrap values
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.utils import resample

# Load a sample dataset (you can replace this with your own dataset)
data = datasets.load_iris()
X, y = data.data, data.target

# Number of bootstrap samples
num_samples = 1000

# Initialize an array to store accuracy scores
accuracy_scores = []

for _ in range(num_samples):
    # Create a bootstrap sample
    X_boot, y_boot = resample(X, y)

    # Split the bootstrap sample into a training and testing set
    X_train, X_test, y_train, y_test = train_test_split(X_boot, y_boot, test_size=0.2, random_state=42)

    # Create an SVM classifier and fit it to the data
    svm_classifier = SVC(kernel='linear')
    svm_classifier.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = svm_classifier.predict(X_test)

    # Calculate the accuracy score
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)

# Calculate the mean and standard deviation of the accuracy scores
mean_accuracy = np.mean(accuracy_scores)
std_accuracy = np.std(accuracy_scores)

# Print the results
print(f"Mean Accuracy: {mean_accuracy}")
print(f"Standard Deviation of Accuracy: {std_accuracy}")
```

```python
#svm bootstrap
import numpy as np
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.utils import resample
import matplotlib.pyplot as plt

# Load a sample dataset (you can replace this with your own dataset)
data = datasets.load_iris()
X, y = data.data, data.target

# Number of bootstrap samples
num_samples = 1000

# Initialize an array to store accuracy scores
accuracy_scores = []

for _ in range(num_samples):
    # Create a bootstrap sample
    X_boot, y_boot = resample(X, y)

    # Create an SVM classifier and fit it to the data
    svm_classifier = SVC(kernel='linear')
    svm_classifier.fit(X_boot, y_boot)

    # Calculate the accuracy score
    accuracy = svm_classifier.score(X, y)  # Here, we calculate the score on the full dataset
    accuracy_scores.append(accuracy)

# Create a histogram of the accuracy scores
plt.hist(accuracy_scores, bins=30, color='skyblue', edgecolor='black')
plt.title('Accuracy Scores Distribution')
plt.xlabel('Accuracy')
plt.ylabel('Frequency')
plt.grid(True)

plt.show()
```

```python
#knn bootstrap
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.utils import resample
import matplotlib.pyplot as plt

# Load a sample dataset (you can replace this with your own dataset)
data = load_iris()
X, y = data.data, data.target

# Number of bootstrap samples
num_samples = 1000

# Number of nearest neighbors for KNN
k_neighbors = 5

# Initialize an array to store accuracy scores
accuracy_scores = []

for _ in range(num_samples):
    # Create a bootstrap sample
    X_boot, y_boot = resample(X, y)

    # Split the bootstrap sample into a training and testing set
    X_train, X_test, y_train, y_test = train_test_split(X_boot, y_boot, test_size=0.2, random_state=42)

    # Create a KNN classifier and fit it to the data
    knn_classifier = KNeighborsClassifier(n_neighbors=k_neighbors)
    knn_classifier.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = knn_classifier.predict(X_test)

    # Calculate the accuracy score
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)

# Calculate the mean and standard deviation of the accuracy scores
mean_accuracy = np.mean(accuracy_scores)
std_accuracy = np.std(accuracy_scores)

# Print the results
print(f"Mean Accuracy: {mean_accuracy}")
print(f"Standard Deviation of Accuracy: {std_accuracy}")

# Create a histogram of the accuracy scores
plt.hist(accuracy_scores, bins=30, color='skyblue', edgecolor='black')
plt.title('Accuracy Scores Distribution')
plt.xlabel('Accuracy')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
